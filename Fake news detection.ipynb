```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import re
import string
```

```python
data_fake = pd.read_csv("Fake.csv", dtype=str)
data_true = pd.read_csv("True.csv")
```

```python
data_fake.head()
```

```python
data_true.head()
```

```python
#added new column
data_fake["class"] = 0
data_true["class"] = 1
```

```python
#deleting 10 rows from bottam from both dataset
data_fake_manual_testing = data_fake.tail(10)
for i in range(23526, 23516,-1):
    data_fake.drop([i], axis = 0 , inplace = True)

data_true_manual_testing = data_true.tail(10)
for i in range(21416, 21406,-1):
    data_true.drop([i], axis = 0 , inplace = True)
```

```python
data_fake.shape, data_true.shape
```

```python
 data_fake_manual_testing["class"] = 0
 data_true_manual_testing["class"] = 1
```

```python
 data_fake_manual_testing.head(10)
```

```python
 data_fake_manual_testing.head(10)
```

```python
#merging both data
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
 data = data_merge.drop(["title","subject","date"],axis = 1)
```

```python
data['text'] = data['text'].fillna(0)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(["index"],axis = 1, inplace =True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    if not isinstance(text, str):
        return ""   # or str(text) if you want to keep numbers
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    return text
```

```python
data["text"] = data["text"].apply(wordopt)
```

```python
x = data["text"]
y = data["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train) #fit the data
xv_test = vectorization.transform(x_test) # Transform the test data
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression() # Train Logistic Regression
LR.fit(xv_train,y_train)
```

```python
# Predict on test set
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test) #cheaking accuracy score
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier() # Instantiate the classifier
DT.fit(xv_train,y_train)
```

```python
# Predict on test set
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test) #cheaking accuracy score
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state=0)  # Instantiate the Gradient Boosting classifier
GB.fit(xv_train, y_train)
```

```python
 # Predict on test set
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test) #cheaking accuracy score
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state=0)  # Instantiate the Random Forest Classifier
RF.fit(xv_train, y_train)
```

```python
# Predict on test set
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test) #cheaking accuracy score
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text": [news]}   # Convert input into a DataFrame
    new_def_test = pd.DataFrame(testing_news)

    new_def_test["text"] = new_def_test["text"].apply(wordopt) # Apply preprocessing
    new_x_test = new_def_test["text"]

    new_xv_test = vectorization.transform(new_x_test) # Vectorize the text

    pred_LR = LR.predict(new_xv_test)[0]  # Predictions from different models
    pred_DT = DT.predict(new_xv_test)[0]
    pred_GB = GB.predict(new_xv_test)[0]
    pred_RF = RF.predict(new_xv_test)[0]


    #Results
    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}" .format(
        output_label(pred_LR),
        output_label(pred_DT),
        output_label(pred_GB),
        output_label(pred_RF),
    ))
```
